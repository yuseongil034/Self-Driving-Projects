```python
# ğŸš— NVIDIA PeopleNet & YOLOv11 ì‚¬ëŒ ì¸ì‹ ì‹œìŠ¤í…œ
# Google Colab í™˜ê²½ì—ì„œ ì•ˆì •ì ìœ¼ë¡œ ì‹¤í–‰ë˜ëŠ” ë²„ì „
# ì´ˆë³´ìë¥¼ ìœ„í•œ ì¹œì ˆí•œ ì½”ë“œ êµ¬ì„±

print("ğŸš— ì‚¬ëŒ ì¸ì‹ ììœ¨ì£¼í–‰ ì‹œìŠ¤í…œì„ ì‹œì‘í•©ë‹ˆë‹¤!")
print("=" * 60)
print("ğŸ“ ì´ ì½”ë“œëŠ” NVIDIA PeopleNetê³¼ YOLOv11ì„ ì‚¬ìš©í•˜ì—¬")
print("   YouTube ì˜ìƒì—ì„œ ì‚¬ëŒì„ ì¸ì‹í•˜ê³  ì‹œê°í™”í•©ë‹ˆë‹¤.")
print("=" * 60)

# ============================================================================
# 1. í™˜ê²½ ì„¤ì • ë° í•„ìˆ˜ ë¼ì´ë¸ŒëŸ¬ë¦¬ ì„¤ì¹˜
# ============================================================================

import subprocess
import sys
import os
import warnings
warnings.filterwarnings('ignore')

def install_and_import(package_name, import_name=None):
    """íŒ¨í‚¤ì§€ ì„¤ì¹˜ ë° ì„í¬íŠ¸ í•¨ìˆ˜"""
    if import_name is None:
        import_name = package_name

    try:
        __import__(import_name)
        print(f"âœ… {package_name} ì´ë¯¸ ì„¤ì¹˜ë¨")
    except ImportError:
        print(f"ğŸ“¦ {package_name} ì„¤ì¹˜ ì¤‘...")
        subprocess.check_call([sys.executable, "-m", "pip", "install", "-q", package_name])
        print(f"âœ… {package_name} ì„¤ì¹˜ ì™„ë£Œ")

# í•„ìˆ˜ ë¼ì´ë¸ŒëŸ¬ë¦¬ ì„¤ì¹˜
print("ğŸ”§ í•„ìš”í•œ ë¼ì´ë¸ŒëŸ¬ë¦¬ë¥¼ ì„¤ì¹˜í•˜ê³  ìˆìŠµë‹ˆë‹¤...")
essential_packages = [
    ("opencv-python", "cv2"),
    ("numpy", "numpy"),
    ("matplotlib", "matplotlib"),
    ("pillow", "PIL"),
    ("ultralytics", "ultralytics"),
    ("yt-dlp", "yt_dlp"),
    ("torch", "torch"),
    ("torchvision", "torchvision")
]

for package, import_name in essential_packages:
    install_and_import(package, import_name)

print("âœ… ëª¨ë“  ë¼ì´ë¸ŒëŸ¬ë¦¬ ì„¤ì¹˜ ì™„ë£Œ!\n")

# ============================================================================
# 2. ë¼ì´ë¸ŒëŸ¬ë¦¬ ì„í¬íŠ¸
# ============================================================================

import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import time
import json
from pathlib import Path
import torch
from ultralytics import YOLO
import yt_dlp
from IPython.display import clear_output, display, HTML
import ipywidgets as widgets

# ============================================================================
# 3. GPU í™˜ê²½ í™•ì¸
# ============================================================================

def check_gpu_environment():
    """GPU ì‚¬ìš© ê°€ëŠ¥ ì—¬ë¶€ í™•ì¸"""
    print("ğŸ” GPU í™˜ê²½ì„ í™•ì¸í•˜ê³  ìˆìŠµë‹ˆë‹¤...")

    # PyTorch CUDA í™•ì¸
    if torch.cuda.is_available():
        gpu_count = torch.cuda.device_count()
        gpu_name = torch.cuda.get_device_name(0)
        print(f"âœ… GPU ì‚¬ìš© ê°€ëŠ¥: {gpu_name}")
        print(f"   GPU ê°œìˆ˜: {gpu_count}")
        print(f"   CUDA ë²„ì „: {torch.version.cuda}")
        return True
    else:
        print("âš ï¸ GPUë¥¼ ì‚¬ìš©í•  ìˆ˜ ì—†ìŠµë‹ˆë‹¤. CPU ëª¨ë“œë¡œ ì‹¤í–‰ë©ë‹ˆë‹¤.")
        print("   (GPU ê°€ì†ì„ ì›í•œë‹¤ë©´ ëŸ°íƒ€ì„ â†’ ëŸ°íƒ€ì„ ìœ í˜• ë³€ê²½ â†’ GPU ì„ íƒ)")
        return False

gpu_available = check_gpu_environment()
device = 'cuda' if gpu_available else 'cpu'
print(f"ğŸ–¥ï¸ ì‚¬ìš© ë””ë°”ì´ìŠ¤: {device.upper()}\n")

# ============================================================================
# 4. ëª¨ë¸ í´ë˜ìŠ¤ ì •ì˜
# ============================================================================

class PeopleDetector:
    """í†µí•© ì‚¬ëŒ ê²€ì¶œ í´ë˜ìŠ¤ (PeopleNet ìŠ¤íƒ€ì¼ + YOLOv11)"""

    def __init__(self, model_type='yolo11', confidence_threshold=0.5):
        """
        ì´ˆê¸°í™” í•¨ìˆ˜
        Args:
            model_type: 'yolo11' ë˜ëŠ” 'yolo8'
            confidence_threshold: ê²€ì¶œ ì‹ ë¢°ë„ ì„ê³„ê°’
        """
        self.model_type = model_type
        self.confidence_threshold = confidence_threshold
        self.model = None
        self.device = device

        print(f"ğŸ¤– {model_type.upper()} ëª¨ë¸ì„ ë¡œë”©í•˜ê³  ìˆìŠµë‹ˆë‹¤...")
        self.load_model()

    def load_model(self):
        """ëª¨ë¸ ë¡œë”©"""
        try:
            if self.model_type == 'yolo11':
                # YOLOv11 ëª¨ë¸ ì‚¬ìš©
                self.model = YOLO('yolo11n.pt')  # nano ë²„ì „ (ë¹ ë¥´ê³  ê°€ë²¼ì›€)
            else:
                # YOLOv8 ëª¨ë¸ ì‚¬ìš© (fallback)
                self.model = YOLO('yolov8n.pt')

            print(f"âœ… {self.model_type.upper()} ëª¨ë¸ ë¡œë”© ì™„ë£Œ")
            print(f"   ë””ë°”ì´ìŠ¤: {self.device}")

        except Exception as e:
            print(f"âŒ ëª¨ë¸ ë¡œë”© ì‹¤íŒ¨: {e}")
            print("   YOLOv8ìœ¼ë¡œ ëŒ€ì²´í•˜ì—¬ ì‹œë„í•©ë‹ˆë‹¤...")
            try:
                self.model = YOLO('yolov8n.pt')
                self.model_type = 'yolo8'
                print("âœ… YOLOv8 ëª¨ë¸ ë¡œë”© ì™„ë£Œ")
            except Exception as e2:
                print(f"âŒ ëª¨ë“  ëª¨ë¸ ë¡œë”© ì‹¤íŒ¨: {e2}")
                self.model = None

    def detect_people(self, image):
        """
        ì´ë¯¸ì§€ì—ì„œ ì‚¬ëŒ ê²€ì¶œ
        Args:
            image: OpenCV ì´ë¯¸ì§€ (BGR)
        Returns:
            detections: [(x1, y1, x2, y2, confidence), ...] í˜•íƒœì˜ ë¦¬ìŠ¤íŠ¸
        """
        if self.model is None:
            return []

        try:
            # YOLO ì¶”ë¡  ì‹¤í–‰
            results = self.model(image, device=self.device, verbose=False)

            detections = []
            for result in results:
                boxes = result.boxes
                if boxes is not None:
                    for box in boxes:
                        # í´ë˜ìŠ¤ ID í™•ì¸ (0 = person)
                        class_id = int(box.cls[0])
                        if class_id == 0:  # person í´ë˜ìŠ¤ë§Œ
                            confidence = float(box.conf[0])
                            if confidence >= self.confidence_threshold:
                                # ì¢Œí‘œ ì¶”ì¶œ (x1, y1, x2, y2)
                                x1, y1, x2, y2 = box.xyxy[0].cpu().numpy()
                                detections.append((int(x1), int(y1), int(x2), int(y2), confidence))

            return detections

        except Exception as e:
            print(f"âŒ ê²€ì¶œ ì¤‘ ì˜¤ë¥˜ ë°œìƒ: {e}")
            return []

# ============================================================================
# 5. YouTube ì˜ìƒ ë‹¤ìš´ë¡œë“œ í•¨ìˆ˜
# ============================================================================

def download_youtube_video():
    """YouTube ì˜ìƒ ë‹¤ìš´ë¡œë“œ"""
    print("ğŸ“º YouTube ì˜ìƒ ë‹¤ìš´ë¡œë“œ")
    print("-" * 40)

    # ì‚¬ìš©ì ì…ë ¥ ë°›ê¸°
    youtube_url = input("ğŸ”— YouTube ë§í¬ë¥¼ ì…ë ¥í•˜ì„¸ìš”: ").strip()
    if not youtube_url:
        print("âŒ ìœ íš¨í•˜ì§€ ì•Šì€ ë§í¬ì…ë‹ˆë‹¤.")
        return None

    output_filename = input("ğŸ’¾ ì €ì¥í•  íŒŒì¼ëª…ì„ ì…ë ¥í•˜ì„¸ìš” (í™•ì¥ì ì œì™¸, ê¸°ë³¸ê°’: video): ").strip()
    if not output_filename:
        output_filename = "video"

    try:
        # yt-dlp ì„¤ì •
        ydl_opts = {
            'format': 'mp4[height<=720]/best[height<=720]',  # 720p ì´í•˜
            'outtmpl': f'{output_filename}.%(ext)s',
            'no_warnings': True,
        }

        print("ğŸ“¥ ì˜ìƒì„ ë‹¤ìš´ë¡œë“œí•˜ê³  ìˆìŠµë‹ˆë‹¤...")
        with yt_dlp.YoutubeDL(ydl_opts) as ydl:
            info = ydl.extract_info(youtube_url, download=False)
            title = info.get('title', 'Unknown')
            duration = info.get('duration', 0)

            print(f"   ğŸ“‹ ì œëª©: {title}")
            print(f"   â±ï¸ ê¸¸ì´: {duration//60}ë¶„ {duration%60}ì´ˆ")

            # ë‹¤ìš´ë¡œë“œ ì‹¤í–‰
            ydl.download([youtube_url])

        # ë‹¤ìš´ë¡œë“œëœ íŒŒì¼ ì°¾ê¸°
        possible_extensions = ['.mp4', '.webm', '.mkv', '.avi']
        video_path = None

        for ext in possible_extensions:
            test_path = f"{output_filename}{ext}"
            if os.path.exists(test_path):
                video_path = test_path
                break

        if video_path and os.path.exists(video_path):
            print(f"âœ… ì˜ìƒ ë‹¤ìš´ë¡œë“œ ì™„ë£Œ: {video_path}")
            return video_path
        else:
            print("âŒ ë‹¤ìš´ë¡œë“œëœ íŒŒì¼ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
            return None

    except Exception as e:
        print(f"âŒ ì˜ìƒ ë‹¤ìš´ë¡œë“œ ì‹¤íŒ¨: {e}")
        return None

# ============================================================================
# 6. ì‹œê°í™” í•¨ìˆ˜
# ============================================================================

def draw_detections(image, detections):
    """
    ê²€ì¶œëœ ì‚¬ëŒì— ë°”ìš´ë”© ë°•ìŠ¤ì™€ ë¼ë²¨ ê·¸ë¦¬ê¸°
    Args:
        image: ì›ë³¸ ì´ë¯¸ì§€
        detections: [(x1, y1, x2, y2, confidence), ...] ë¦¬ìŠ¤íŠ¸
    Returns:
        annotated_image: ì£¼ì„ì´ ì¶”ê°€ëœ ì´ë¯¸ì§€
    """
    annotated_image = image.copy()

    # ìƒ‰ìƒ ì •ì˜ (BGR)
    box_color = (0, 255, 0)  # ì´ˆë¡ìƒ‰
    text_color = (255, 255, 255)  # í°ìƒ‰
    bg_color = (0, 200, 0)  # ì§„í•œ ì´ˆë¡ìƒ‰

    for detection in detections:
        x1, y1, x2, y2, confidence = detection

        # ë°”ìš´ë”© ë°•ìŠ¤ ê·¸ë¦¬ê¸°
        cv2.rectangle(annotated_image, (x1, y1), (x2, y2), box_color, 2)

        # ë¼ë²¨ í…ìŠ¤íŠ¸ ì¤€ë¹„
        label = f"Person: {confidence:.2f}"

        # í…ìŠ¤íŠ¸ í¬ê¸° ê³„ì‚°
        font = cv2.FONT_HERSHEY_SIMPLEX
        font_scale = 0.6
        thickness = 1
        (text_width, text_height), baseline = cv2.getTextSize(label, font, font_scale, thickness)

        # ë¼ë²¨ ë°°ê²½ ê·¸ë¦¬ê¸°
        cv2.rectangle(
            annotated_image,
            (x1, y1 - text_height - baseline - 10),
            (x1 + text_width + 10, y1),
            bg_color,
            cv2.FILLED
        )

        # ë¼ë²¨ í…ìŠ¤íŠ¸ ê·¸ë¦¬ê¸°
        cv2.putText(
            annotated_image,
            label,
            (x1 + 5, y1 - baseline - 5),
            font,
            font_scale,
            text_color,
            thickness
        )

    return annotated_image

def show_detection_results(original_image, detections, title="ê²€ì¶œ ê²°ê³¼"):
    """ê²€ì¶œ ê²°ê³¼ë¥¼ matplotlibìœ¼ë¡œ í‘œì‹œ"""
    annotated_image = draw_detections(original_image, detections)

    # BGRì„ RGBë¡œ ë³€í™˜
    original_rgb = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)
    annotated_rgb = cv2.cvtColor(annotated_image, cv2.COLOR_BGR2RGB)

    # í”Œë¡¯ ìƒì„±
    fig, axes = plt.subplots(1, 2, figsize=(15, 8))

    axes[0].imshow(original_rgb)
    axes[0].set_title("ì›ë³¸ ì´ë¯¸ì§€", fontsize=14)
    axes[0].axis('off')

    axes[1].imshow(annotated_rgb)
    axes[1].set_title(f"{title} ({len(detections)}ëª… ê²€ì¶œ)", fontsize=14)
    axes[1].axis('off')

    plt.tight_layout()
    plt.show()

# ============================================================================
# 7. ì˜ìƒ ì²˜ë¦¬ í•¨ìˆ˜
# ============================================================================

def process_video(detector, video_path, output_path=None, show_progress=True):
    """
    ì˜ìƒì—ì„œ ì‚¬ëŒ ê²€ì¶œ ë° ì²˜ë¦¬
    Args:
        detector: PeopleDetector ì¸ìŠ¤í„´ìŠ¤
        video_path: ì…ë ¥ ì˜ìƒ ê²½ë¡œ
        output_path: ì¶œë ¥ ì˜ìƒ ê²½ë¡œ (Noneì´ë©´ ì €ì¥ ì•ˆí•¨)
        show_progress: ì§„í–‰ë¥  í‘œì‹œ ì—¬ë¶€
    """
    print(f"ğŸ¥ ì˜ìƒ ì²˜ë¦¬ ì‹œì‘: {video_path}")

    # ì˜ìƒ ì—´ê¸°
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        print("âŒ ì˜ìƒì„ ì—´ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
        return None

    # ì˜ìƒ ì •ë³´ ê°€ì ¸ì˜¤ê¸°
    fps = int(cap.get(cv2.CAP_PROP_FPS))
    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    duration = total_frames / fps

    print(f"ğŸ“Š ì˜ìƒ ì •ë³´:")
    print(f"   í•´ìƒë„: {width}x{height}")
    print(f"   FPS: {fps}")
    print(f"   ì´ í”„ë ˆì„: {total_frames}")
    print(f"   ê¸¸ì´: {duration:.1f}ì´ˆ")

    # ì¶œë ¥ ì˜ìƒ ì„¤ì •
    writer = None
    if output_path:
        fourcc = cv2.VideoWriter_fourcc(*'mp4v')
        writer = cv2.VideoWriter(output_path, fourcc, fps, (width, height))
        print(f"ğŸ’¾ ê²°ê³¼ ì˜ìƒ ì €ì¥ ê²½ë¡œ: {output_path}")

    # í†µê³„ ë³€ìˆ˜
    frame_count = 0
    detection_stats = []
    processing_times = []

    # ìƒ˜í”Œ í”„ë ˆì„ ì €ì¥ìš©
    sample_frames = []
    sample_detections = []

    try:
        print("\nğŸ”„ ì˜ìƒ ì²˜ë¦¬ ì¤‘...")
        start_time = time.time()

        while True:
            ret, frame = cap.read()
            if not ret:
                break

            frame_count += 1

            # ì²˜ë¦¬ ì‹œê°„ ì¸¡ì •
            frame_start = time.time()
            detections = detector.detect_people(frame)
            frame_end = time.time()

            processing_times.append(frame_end - frame_start)
            detection_stats.append(len(detections))

            # ê²€ì¶œ ê²°ê³¼ ê·¸ë¦¬ê¸°
            annotated_frame = draw_detections(frame, detections)

            # í”„ë ˆì„ ì •ë³´ ì¶”ê°€
            info_text = f"Frame: {frame_count}/{total_frames} | People: {len(detections)} | FPS: {1/(frame_end-frame_start):.1f}"
            cv2.putText(annotated_frame, info_text, (10, 30),
                       cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 2)

            # ì˜ìƒ ì €ì¥
            if writer:
                writer.write(annotated_frame)

            # ìƒ˜í”Œ í”„ë ˆì„ ì €ì¥ (ì²˜ìŒ, ì¤‘ê°„, ë ë¶€ë¶„)
            if frame_count in [1, total_frames//2, total_frames-1]:
                sample_frames.append(frame.copy())
                sample_detections.append(detections)

            # ì§„í–‰ë¥  í‘œì‹œ
            if show_progress and frame_count % max(1, total_frames//50) == 0:
                progress = (frame_count / total_frames) * 100
                elapsed = time.time() - start_time
                eta = elapsed * (total_frames - frame_count) / frame_count if frame_count > 0 else 0
                print(f"   ì§„í–‰ë¥ : {progress:.1f}% | ë‚¨ì€ì‹œê°„: {eta:.1f}ì´ˆ", end='\r')

        total_time = time.time() - start_time

    except KeyboardInterrupt:
        print("\nâ¹ï¸ ì‚¬ìš©ìì— ì˜í•´ ì¤‘ë‹¨ë˜ì—ˆìŠµë‹ˆë‹¤.")

    finally:
        cap.release()
        if writer:
            writer.release()

    # ê²°ê³¼ ì¶œë ¥
    print(f"\nâœ… ì˜ìƒ ì²˜ë¦¬ ì™„ë£Œ!")
    print(f"ğŸ“Š ì²˜ë¦¬ í†µê³„:")
    print(f"   ì´ ì²˜ë¦¬ ì‹œê°„: {total_time:.1f}ì´ˆ")
    print(f"   í‰ê·  FPS: {frame_count/total_time:.1f}")
    print(f"   í‰ê·  ê²€ì¶œ ì¸ì›: {np.mean(detection_stats):.1f}ëª…")
    print(f"   ìµœëŒ€ ê²€ì¶œ ì¸ì›: {np.max(detection_stats)}ëª…")
    print(f"   í‰ê·  í”„ë ˆì„ ì²˜ë¦¬ ì‹œê°„: {np.mean(processing_times)*1000:.1f}ms")

    # ìƒ˜í”Œ í”„ë ˆì„ ì‹œê°í™”
    if sample_frames:
        print("\nğŸ–¼ï¸ ìƒ˜í”Œ í”„ë ˆì„ ê²°ê³¼:")
        for i, (frame, detections) in enumerate(zip(sample_frames, sample_detections)):
            position = ["ì‹œì‘", "ì¤‘ê°„", "ë"][i]
            show_detection_results(frame, detections, f"{position} ë¶€ë¶„")

    return {
        'total_frames': frame_count,
        'total_time': total_time,
        'detection_stats': detection_stats,
        'processing_times': processing_times,
        'output_path': output_path
    }

# ============================================================================
# 8. ë©”ì¸ ì‹¤í–‰ í•¨ìˆ˜
# ============================================================================

def main():
    """ë©”ì¸ ì‹¤í–‰ í•¨ìˆ˜"""
    print("ğŸš€ ì‚¬ëŒ ì¸ì‹ ì‹œìŠ¤í…œì„ ì‹œì‘í•©ë‹ˆë‹¤!")
    print("=" * 50)

    # ëª¨ë¸ ì´ˆê¸°í™”
    print("1ï¸âƒ£ ëª¨ë¸ì„ ì´ˆê¸°í™”í•˜ê³  ìˆìŠµë‹ˆë‹¤...")
    detector = PeopleDetector(model_type='yolo11', confidence_threshold=0.5)

    if detector.model is None:
        print("âŒ ëª¨ë¸ ì´ˆê¸°í™”ì— ì‹¤íŒ¨í–ˆìŠµë‹ˆë‹¤. í”„ë¡œê·¸ë¨ì„ ì¢…ë£Œí•©ë‹ˆë‹¤.")
        return

    print("2ï¸âƒ£ YouTube ì˜ìƒì„ ë‹¤ìš´ë¡œë“œí•©ë‹ˆë‹¤...")
    video_path = download_youtube_video()

    if video_path is None:
        print("âŒ ì˜ìƒ ë‹¤ìš´ë¡œë“œì— ì‹¤íŒ¨í–ˆìŠµë‹ˆë‹¤.")
        return

    print("3ï¸âƒ£ ê²°ê³¼ ì˜ìƒ ì €ì¥ ì„¤ì •...")
    save_result = input("ğŸ¬ ì²˜ë¦¬ëœ ì˜ìƒì„ ì €ì¥í•˜ì‹œê² ìŠµë‹ˆê¹Œ? (y/n, ê¸°ë³¸ê°’: y): ").strip().lower()

    output_path = None
    if save_result != 'n':
        output_filename = input("ğŸ’¾ ì €ì¥í•  íŒŒì¼ëª…ì„ ì…ë ¥í•˜ì„¸ìš” (í™•ì¥ì ì œì™¸, ê¸°ë³¸ê°’: processed_video): ").strip()
        if not output_filename:
            output_filename = "processed_video"
        output_path = f"{output_filename}.mp4"

    print("4ï¸âƒ£ ì˜ìƒ ì²˜ë¦¬ë¥¼ ì‹œì‘í•©ë‹ˆë‹¤!")
    results = process_video(detector, video_path, output_path)

    if results:
        print("\nğŸ‰ ëª¨ë“  ì²˜ë¦¬ê°€ ì™„ë£Œë˜ì—ˆìŠµë‹ˆë‹¤!")
        if output_path and os.path.exists(output_path):
            print(f"âœ… ê²°ê³¼ ì˜ìƒì´ ì €ì¥ë˜ì—ˆìŠµë‹ˆë‹¤: {output_path}")

    print("\nğŸ“‹ ììœ¨ì£¼í–‰ì—ì„œì˜ í™œìš©:")
    print("   â€¢ íš¡ë‹¨ë³´ë„ ë³´í–‰ì ê°ì§€")
    print("   â€¢ ìŠ¤ì¿¨ì¡´ ì–´ë¦°ì´ ë³´í˜¸")
    print("   â€¢ ì£¼ì°¨ì¥ ë³´í–‰ì ì•ˆì „")
    print("   â€¢ ì‘ê¸‰ìƒí™© ëŒ€ì‘")

# ============================================================================
# 9. ì‹¤í–‰ ì½”ë“œ
# ============================================================================

if __name__ == "__main__":
    try:
        main()
    except KeyboardInterrupt:
        print("\nğŸ‘‹ í”„ë¡œê·¸ë¨ì´ ì‚¬ìš©ìì— ì˜í•´ ì¢…ë£Œë˜ì—ˆìŠµë‹ˆë‹¤.")
    except Exception as e:
        print(f"\nâŒ ì˜¤ë¥˜ê°€ ë°œìƒí–ˆìŠµë‹ˆë‹¤: {e}")
        print("ğŸ”§ ë¬¸ì œê°€ ì§€ì†ë˜ë©´ ë‹¤ì‹œ ì‹¤í–‰í•´ë³´ì„¸ìš”.")

print(f"""
{'='*60}
ğŸš— NVIDIA PeopleNet & YOLOv11 ì‚¬ëŒ ì¸ì‹ ì‹œìŠ¤í…œ
{'='*60}
âœ… Google Colab ìµœì í™” ì™„ë£Œ
âœ… ì´ˆë³´ì ì¹œí™”ì  ì½”ë“œ êµ¬ì„±
âœ… YouTube ì˜ìƒ ìë™ ë‹¤ìš´ë¡œë“œ
âœ… ì‹¤ì‹œê°„ ì§„í–‰ë¥  í‘œì‹œ
âœ… ìë™ GPU/CPU ê°ì§€
{'='*60}
""")

# ì‚¬ìš©ë²• ì•ˆë‚´
print("ğŸ¯ ì‚¬ìš©ë²•:")
print("1. ìœ„ ì½”ë“œë¥¼ ì‹¤í–‰í•˜ì„¸ìš”")
print("2. YouTube ë§í¬ë¥¼ ì…ë ¥í•˜ì„¸ìš”")
print("3. íŒŒì¼ëª…ì„ ì„¤ì •í•˜ì„¸ìš”")
print("4. ê²°ê³¼ë¥¼ í™•ì¸í•˜ì„¸ìš”!")
print("\nâš¡ íŒ: GPU ëŸ°íƒ€ì„ ì‚¬ìš© ì‹œ ë” ë¹ ë¥¸ ì²˜ë¦¬ê°€ ê°€ëŠ¥í•©ë‹ˆë‹¤!")
```

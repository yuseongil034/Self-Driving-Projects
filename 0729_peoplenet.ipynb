```python
# 🚗 NVIDIA PeopleNet & YOLOv11 사람 인식 시스템
# Google Colab 환경에서 안정적으로 실행되는 버전
# 초보자를 위한 친절한 코드 구성

print("🚗 사람 인식 자율주행 시스템을 시작합니다!")
print("=" * 60)
print("📝 이 코드는 NVIDIA PeopleNet과 YOLOv11을 사용하여")
print("   YouTube 영상에서 사람을 인식하고 시각화합니다.")
print("=" * 60)

# ============================================================================
# 1. 환경 설정 및 필수 라이브러리 설치
# ============================================================================

import subprocess
import sys
import os
import warnings
warnings.filterwarnings('ignore')

def install_and_import(package_name, import_name=None):
    """패키지 설치 및 임포트 함수"""
    if import_name is None:
        import_name = package_name

    try:
        __import__(import_name)
        print(f"✅ {package_name} 이미 설치됨")
    except ImportError:
        print(f"📦 {package_name} 설치 중...")
        subprocess.check_call([sys.executable, "-m", "pip", "install", "-q", package_name])
        print(f"✅ {package_name} 설치 완료")

# 필수 라이브러리 설치
print("🔧 필요한 라이브러리를 설치하고 있습니다...")
essential_packages = [
    ("opencv-python", "cv2"),
    ("numpy", "numpy"),
    ("matplotlib", "matplotlib"),
    ("pillow", "PIL"),
    ("ultralytics", "ultralytics"),
    ("yt-dlp", "yt_dlp"),
    ("torch", "torch"),
    ("torchvision", "torchvision")
]

for package, import_name in essential_packages:
    install_and_import(package, import_name)

print("✅ 모든 라이브러리 설치 완료!\n")

# ============================================================================
# 2. 라이브러리 임포트
# ============================================================================

import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import time
import json
from pathlib import Path
import torch
from ultralytics import YOLO
import yt_dlp
from IPython.display import clear_output, display, HTML
import ipywidgets as widgets

# ============================================================================
# 3. GPU 환경 확인
# ============================================================================

def check_gpu_environment():
    """GPU 사용 가능 여부 확인"""
    print("🔍 GPU 환경을 확인하고 있습니다...")

    # PyTorch CUDA 확인
    if torch.cuda.is_available():
        gpu_count = torch.cuda.device_count()
        gpu_name = torch.cuda.get_device_name(0)
        print(f"✅ GPU 사용 가능: {gpu_name}")
        print(f"   GPU 개수: {gpu_count}")
        print(f"   CUDA 버전: {torch.version.cuda}")
        return True
    else:
        print("⚠️ GPU를 사용할 수 없습니다. CPU 모드로 실행됩니다.")
        print("   (GPU 가속을 원한다면 런타임 → 런타임 유형 변경 → GPU 선택)")
        return False

gpu_available = check_gpu_environment()
device = 'cuda' if gpu_available else 'cpu'
print(f"🖥️ 사용 디바이스: {device.upper()}\n")

# ============================================================================
# 4. 모델 클래스 정의
# ============================================================================

class PeopleDetector:
    """통합 사람 검출 클래스 (PeopleNet 스타일 + YOLOv11)"""

    def __init__(self, model_type='yolo11', confidence_threshold=0.5):
        """
        초기화 함수
        Args:
            model_type: 'yolo11' 또는 'yolo8'
            confidence_threshold: 검출 신뢰도 임계값
        """
        self.model_type = model_type
        self.confidence_threshold = confidence_threshold
        self.model = None
        self.device = device

        print(f"🤖 {model_type.upper()} 모델을 로딩하고 있습니다...")
        self.load_model()

    def load_model(self):
        """모델 로딩"""
        try:
            if self.model_type == 'yolo11':
                # YOLOv11 모델 사용
                self.model = YOLO('yolo11n.pt')  # nano 버전 (빠르고 가벼움)
            else:
                # YOLOv8 모델 사용 (fallback)
                self.model = YOLO('yolov8n.pt')

            print(f"✅ {self.model_type.upper()} 모델 로딩 완료")
            print(f"   디바이스: {self.device}")

        except Exception as e:
            print(f"❌ 모델 로딩 실패: {e}")
            print("   YOLOv8으로 대체하여 시도합니다...")
            try:
                self.model = YOLO('yolov8n.pt')
                self.model_type = 'yolo8'
                print("✅ YOLOv8 모델 로딩 완료")
            except Exception as e2:
                print(f"❌ 모든 모델 로딩 실패: {e2}")
                self.model = None

    def detect_people(self, image):
        """
        이미지에서 사람 검출
        Args:
            image: OpenCV 이미지 (BGR)
        Returns:
            detections: [(x1, y1, x2, y2, confidence), ...] 형태의 리스트
        """
        if self.model is None:
            return []

        try:
            # YOLO 추론 실행
            results = self.model(image, device=self.device, verbose=False)

            detections = []
            for result in results:
                boxes = result.boxes
                if boxes is not None:
                    for box in boxes:
                        # 클래스 ID 확인 (0 = person)
                        class_id = int(box.cls[0])
                        if class_id == 0:  # person 클래스만
                            confidence = float(box.conf[0])
                            if confidence >= self.confidence_threshold:
                                # 좌표 추출 (x1, y1, x2, y2)
                                x1, y1, x2, y2 = box.xyxy[0].cpu().numpy()
                                detections.append((int(x1), int(y1), int(x2), int(y2), confidence))

            return detections

        except Exception as e:
            print(f"❌ 검출 중 오류 발생: {e}")
            return []

# ============================================================================
# 5. YouTube 영상 다운로드 함수
# ============================================================================

def download_youtube_video():
    """YouTube 영상 다운로드"""
    print("📺 YouTube 영상 다운로드")
    print("-" * 40)

    # 사용자 입력 받기
    youtube_url = input("🔗 YouTube 링크를 입력하세요: ").strip()
    if not youtube_url:
        print("❌ 유효하지 않은 링크입니다.")
        return None

    output_filename = input("💾 저장할 파일명을 입력하세요 (확장자 제외, 기본값: video): ").strip()
    if not output_filename:
        output_filename = "video"

    try:
        # yt-dlp 설정
        ydl_opts = {
            'format': 'mp4[height<=720]/best[height<=720]',  # 720p 이하
            'outtmpl': f'{output_filename}.%(ext)s',
            'no_warnings': True,
        }

        print("📥 영상을 다운로드하고 있습니다...")
        with yt_dlp.YoutubeDL(ydl_opts) as ydl:
            info = ydl.extract_info(youtube_url, download=False)
            title = info.get('title', 'Unknown')
            duration = info.get('duration', 0)

            print(f"   📋 제목: {title}")
            print(f"   ⏱️ 길이: {duration//60}분 {duration%60}초")

            # 다운로드 실행
            ydl.download([youtube_url])

        # 다운로드된 파일 찾기
        possible_extensions = ['.mp4', '.webm', '.mkv', '.avi']
        video_path = None

        for ext in possible_extensions:
            test_path = f"{output_filename}{ext}"
            if os.path.exists(test_path):
                video_path = test_path
                break

        if video_path and os.path.exists(video_path):
            print(f"✅ 영상 다운로드 완료: {video_path}")
            return video_path
        else:
            print("❌ 다운로드된 파일을 찾을 수 없습니다.")
            return None

    except Exception as e:
        print(f"❌ 영상 다운로드 실패: {e}")
        return None

# ============================================================================
# 6. 시각화 함수
# ============================================================================

def draw_detections(image, detections):
    """
    검출된 사람에 바운딩 박스와 라벨 그리기
    Args:
        image: 원본 이미지
        detections: [(x1, y1, x2, y2, confidence), ...] 리스트
    Returns:
        annotated_image: 주석이 추가된 이미지
    """
    annotated_image = image.copy()

    # 색상 정의 (BGR)
    box_color = (0, 255, 0)  # 초록색
    text_color = (255, 255, 255)  # 흰색
    bg_color = (0, 200, 0)  # 진한 초록색

    for detection in detections:
        x1, y1, x2, y2, confidence = detection

        # 바운딩 박스 그리기
        cv2.rectangle(annotated_image, (x1, y1), (x2, y2), box_color, 2)

        # 라벨 텍스트 준비
        label = f"Person: {confidence:.2f}"

        # 텍스트 크기 계산
        font = cv2.FONT_HERSHEY_SIMPLEX
        font_scale = 0.6
        thickness = 1
        (text_width, text_height), baseline = cv2.getTextSize(label, font, font_scale, thickness)

        # 라벨 배경 그리기
        cv2.rectangle(
            annotated_image,
            (x1, y1 - text_height - baseline - 10),
            (x1 + text_width + 10, y1),
            bg_color,
            cv2.FILLED
        )

        # 라벨 텍스트 그리기
        cv2.putText(
            annotated_image,
            label,
            (x1 + 5, y1 - baseline - 5),
            font,
            font_scale,
            text_color,
            thickness
        )

    return annotated_image

def show_detection_results(original_image, detections, title="검출 결과"):
    """검출 결과를 matplotlib으로 표시"""
    annotated_image = draw_detections(original_image, detections)

    # BGR을 RGB로 변환
    original_rgb = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)
    annotated_rgb = cv2.cvtColor(annotated_image, cv2.COLOR_BGR2RGB)

    # 플롯 생성
    fig, axes = plt.subplots(1, 2, figsize=(15, 8))

    axes[0].imshow(original_rgb)
    axes[0].set_title("원본 이미지", fontsize=14)
    axes[0].axis('off')

    axes[1].imshow(annotated_rgb)
    axes[1].set_title(f"{title} ({len(detections)}명 검출)", fontsize=14)
    axes[1].axis('off')

    plt.tight_layout()
    plt.show()

# ============================================================================
# 7. 영상 처리 함수
# ============================================================================

def process_video(detector, video_path, output_path=None, show_progress=True):
    """
    영상에서 사람 검출 및 처리
    Args:
        detector: PeopleDetector 인스턴스
        video_path: 입력 영상 경로
        output_path: 출력 영상 경로 (None이면 저장 안함)
        show_progress: 진행률 표시 여부
    """
    print(f"🎥 영상 처리 시작: {video_path}")

    # 영상 열기
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        print("❌ 영상을 열 수 없습니다.")
        return None

    # 영상 정보 가져오기
    fps = int(cap.get(cv2.CAP_PROP_FPS))
    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    duration = total_frames / fps

    print(f"📊 영상 정보:")
    print(f"   해상도: {width}x{height}")
    print(f"   FPS: {fps}")
    print(f"   총 프레임: {total_frames}")
    print(f"   길이: {duration:.1f}초")

    # 출력 영상 설정
    writer = None
    if output_path:
        fourcc = cv2.VideoWriter_fourcc(*'mp4v')
        writer = cv2.VideoWriter(output_path, fourcc, fps, (width, height))
        print(f"💾 결과 영상 저장 경로: {output_path}")

    # 통계 변수
    frame_count = 0
    detection_stats = []
    processing_times = []

    # 샘플 프레임 저장용
    sample_frames = []
    sample_detections = []

    try:
        print("\n🔄 영상 처리 중...")
        start_time = time.time()

        while True:
            ret, frame = cap.read()
            if not ret:
                break

            frame_count += 1

            # 처리 시간 측정
            frame_start = time.time()
            detections = detector.detect_people(frame)
            frame_end = time.time()

            processing_times.append(frame_end - frame_start)
            detection_stats.append(len(detections))

            # 검출 결과 그리기
            annotated_frame = draw_detections(frame, detections)

            # 프레임 정보 추가
            info_text = f"Frame: {frame_count}/{total_frames} | People: {len(detections)} | FPS: {1/(frame_end-frame_start):.1f}"
            cv2.putText(annotated_frame, info_text, (10, 30),
                       cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 2)

            # 영상 저장
            if writer:
                writer.write(annotated_frame)

            # 샘플 프레임 저장 (처음, 중간, 끝 부분)
            if frame_count in [1, total_frames//2, total_frames-1]:
                sample_frames.append(frame.copy())
                sample_detections.append(detections)

            # 진행률 표시
            if show_progress and frame_count % max(1, total_frames//50) == 0:
                progress = (frame_count / total_frames) * 100
                elapsed = time.time() - start_time
                eta = elapsed * (total_frames - frame_count) / frame_count if frame_count > 0 else 0
                print(f"   진행률: {progress:.1f}% | 남은시간: {eta:.1f}초", end='\r')

        total_time = time.time() - start_time

    except KeyboardInterrupt:
        print("\n⏹️ 사용자에 의해 중단되었습니다.")

    finally:
        cap.release()
        if writer:
            writer.release()

    # 결과 출력
    print(f"\n✅ 영상 처리 완료!")
    print(f"📊 처리 통계:")
    print(f"   총 처리 시간: {total_time:.1f}초")
    print(f"   평균 FPS: {frame_count/total_time:.1f}")
    print(f"   평균 검출 인원: {np.mean(detection_stats):.1f}명")
    print(f"   최대 검출 인원: {np.max(detection_stats)}명")
    print(f"   평균 프레임 처리 시간: {np.mean(processing_times)*1000:.1f}ms")

    # 샘플 프레임 시각화
    if sample_frames:
        print("\n🖼️ 샘플 프레임 결과:")
        for i, (frame, detections) in enumerate(zip(sample_frames, sample_detections)):
            position = ["시작", "중간", "끝"][i]
            show_detection_results(frame, detections, f"{position} 부분")

    return {
        'total_frames': frame_count,
        'total_time': total_time,
        'detection_stats': detection_stats,
        'processing_times': processing_times,
        'output_path': output_path
    }

# ============================================================================
# 8. 메인 실행 함수
# ============================================================================

def main():
    """메인 실행 함수"""
    print("🚀 사람 인식 시스템을 시작합니다!")
    print("=" * 50)

    # 모델 초기화
    print("1️⃣ 모델을 초기화하고 있습니다...")
    detector = PeopleDetector(model_type='yolo11', confidence_threshold=0.5)

    if detector.model is None:
        print("❌ 모델 초기화에 실패했습니다. 프로그램을 종료합니다.")
        return

    print("2️⃣ YouTube 영상을 다운로드합니다...")
    video_path = download_youtube_video()

    if video_path is None:
        print("❌ 영상 다운로드에 실패했습니다.")
        return

    print("3️⃣ 결과 영상 저장 설정...")
    save_result = input("🎬 처리된 영상을 저장하시겠습니까? (y/n, 기본값: y): ").strip().lower()

    output_path = None
    if save_result != 'n':
        output_filename = input("💾 저장할 파일명을 입력하세요 (확장자 제외, 기본값: processed_video): ").strip()
        if not output_filename:
            output_filename = "processed_video"
        output_path = f"{output_filename}.mp4"

    print("4️⃣ 영상 처리를 시작합니다!")
    results = process_video(detector, video_path, output_path)

    if results:
        print("\n🎉 모든 처리가 완료되었습니다!")
        if output_path and os.path.exists(output_path):
            print(f"✅ 결과 영상이 저장되었습니다: {output_path}")

    print("\n📋 자율주행에서의 활용:")
    print("   • 횡단보도 보행자 감지")
    print("   • 스쿨존 어린이 보호")
    print("   • 주차장 보행자 안전")
    print("   • 응급상황 대응")

# ============================================================================
# 9. 실행 코드
# ============================================================================

if __name__ == "__main__":
    try:
        main()
    except KeyboardInterrupt:
        print("\n👋 프로그램이 사용자에 의해 종료되었습니다.")
    except Exception as e:
        print(f"\n❌ 오류가 발생했습니다: {e}")
        print("🔧 문제가 지속되면 다시 실행해보세요.")

print(f"""
{'='*60}
🚗 NVIDIA PeopleNet & YOLOv11 사람 인식 시스템
{'='*60}
✅ Google Colab 최적화 완료
✅ 초보자 친화적 코드 구성
✅ YouTube 영상 자동 다운로드
✅ 실시간 진행률 표시
✅ 자동 GPU/CPU 감지
{'='*60}
""")

# 사용법 안내
print("🎯 사용법:")
print("1. 위 코드를 실행하세요")
print("2. YouTube 링크를 입력하세요")
print("3. 파일명을 설정하세요")
print("4. 결과를 확인하세요!")
print("\n⚡ 팁: GPU 런타임 사용 시 더 빠른 처리가 가능합니다!")
```
